```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelBinarizer
```

```python
iris = load_iris()
```

```python
pd.DataFrame(iris.data, columns=iris.feature_names).head()
```

```python
pd.DataFrame(iris.target).replace(0, iris.target_names[0]).replace(1, iris.target_names[1]).replace(2, iris.target_names[2]).head()
```

```python
lb = LabelBinarizer()
lb.fit(iris.target)
temp = lb.transform(iris.target)

xtrain, xtest, ytrain, ytest = train_test_split(iris.data, temp, test_size=0.2)
```

```python
X = tf.placeholder('float', [None, 4])
y = tf.placeholder('float')
```

```python
hsize = 16
n_classes = 3
n_features = 4

def neural_net_model(data):
    hl1 = {'weights': tf.Variable(tf.random_normal([n_features, hsize])),
           'biases': tf.Variable(tf.zeros([hsize]))}
    
    hl2 = {'weights': tf.Variable(tf.random_normal([hsize, hsize])),
           'biases': tf.Variable(tf.zeros([hsize]))}
    
    out = {'weights': tf.Variable(tf.random_normal([hsize, n_classes])),
           'biases': tf.Variable(tf.zeros([n_classes]))}
    
    l = tf.add(tf.matmul(data, hl1['weights']), hl1['biases'])
    l = tf.nn.relu(l)
    
    l2 = tf.add(tf.matmul(l, hl2['weights']), hl2['biases'])
    l2 = tf.nn.relu(l2)
    
    o = tf.add(tf.matmul(l2, out['weights']), out['biases'])
    return o
```

```python
def train_neural_network(x):
    prediction = neural_net_model(x)
    cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=prediction, labels=y))
    optimizer = tf.train.AdamOptimizer().minimize(cost)
    
    hm_ep = 1000
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        
        for ep in range(hm_ep):
            _, c = sess.run([optimizer, cost], feed_dict={X: xtrain, y: ytrain})
            print('{} loss: {}'.format(ep, c))
            
        correct = tf.equal(tf.argmax(prediction, 1), tf.argmax(y, 1))
        accuracy = tf.reduce_mean(tf.cast(correct, 'float'))
        print('Accuracy: {}'.format(accuracy.eval({X: xtest, y: ytest})))
```

```python
train_neural_network(X)
```
